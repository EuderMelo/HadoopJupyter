# Sqoop

- https://sqoop.apache.org/

## Setup

- download from https://downloads.apache.org/sqoop/1.4.7
- version 1.4.7

```python
%%bash

# Download package
wget -q -c https://downloads.apache.org/sqoop/1.4.7/sqoop-1.4.7.bin__hadoop-2.6.0.tar.gz

# Copy installation package to container
docker cp sqoop-1.4.7.bin__hadoop-2.6.0.tar.gz hadoop:/opt
```

```python
%%dockerexec -u hadoop hadoop

# unpack file and create link
tar -zxf /opt/sqoop-1.4.7.bin__hadoop-2.6.0.tar.gz -C /opt
ln -s /opt/sqoop-1.4.7.bin__hadoop-2.6.0 /opt/sqoop

# update commons-lang
rm /opt/sqoop/lib/commons-lang3-3.4.jar
cp /opt/hadoop/share/hadoop/yarn/timelineservice/lib/commons-lang-2.6.jar /opt/sqoop/lib

# update envvars.sh
cat >> /opt/envvars.sh << EOF
# Sqoop
export SQOOP_HOME=/opt/sqoop
export PATH=\$PATH:\$SQOOP_HOME/bin

EOF

sudo rm /opt/sqoop-1.4.7.bin__hadoop-2.6.0.tar.gz
```

### Mysql-connector

- https://dev.mysql.com/downloads/connector/j/

```python
%%bash

# Download package
wget -q -c https://dev.mysql.com/get/Downloads/Connector-J/mysql-connector-java_8.0.22-1ubuntu18.04_all.deb

# Copy installation package to container
docker cp mysql-connector-java_8.0.22-1ubuntu18.04_all.deb hadoop:/tmp
    
docker exec hadoop dpkg -i /tmp/mysql-connector-java_8.0.22-1ubuntu18.04_all.deb
```

```python
%%dockerexec -u hadoop hadoop

cp /usr/share/java/mysql-connector-java-8.0.22.jar /opt/sqoop/lib
```

## Mysql installation

```python
%%dockerexec hadoop

apt install -qq -y mysql-server unzip >> install.log 2>&1

# Enable external access (from worker nodes)
sed -i "s/^bind-address/#bind-address/g" /etc/mysql/mysql.conf.d/mysqld.cnf 

service mysql restart
service mysql status
```

```python
%%dockerexec hadoop

# create hadoop user
mysql -e "create user 'hadoop'"
mysql -e "grant all privileges on *.* to 'hadoop'@'%'"
mysql -e "flush privileges"
```

## Employees database setup

```python
%%bash

# Download EmployeesDB sample database
wget -q -c https://github.com/datacharmer/test_db/archive/master.zip

docker cp master.zip hadoop:/opt
```

```python
%%dockerexec -u hadoop hadoop

cd /opt
unzip master.zip
cd test_db-master
mysql -u hadoop < employees.sql
```

```python
%%dockerexec -u hadoop hadoop

mysql -u hadoop -e 'show databases'

printf "\n%40s\n\n" | tr ' ' '='

mysql -u hadoop -D employees -e 'show tables'

printf "\n%40s\n\n" | tr ' ' '='

mysql -u hadoop -D employees -e 'describe employees'
```

## Using sqoop

```python
%%dockerexec -u hadoop hadoop
source /opt/envvars.sh

$SQOOP_HOME/bin/sqoop list-databases --connect jdbc:mysql://hadoop --username hadoop
```

```python
%%dockerexec -u hadoop hadoop
source /opt/envvars.sh

$SQOOP_HOME/bin/sqoop list-tables --connect jdbc:mysql://hadoop/employees --username hadoop
```

```python
%%dockerexec -u hadoop hadoop
source /opt/envvars.sh

$SQOOP_HOME/bin/sqoop import --connect jdbc:mysql://hadoop/employees --username hadoop --table employees
```

```python
%%dockerexec -u hadoop hadoop
source /opt/envvars.sh

hdfs dfs -ls -h employees

hdfs dfs -head employees/part-m-00000
```
